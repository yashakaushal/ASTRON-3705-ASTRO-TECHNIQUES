### Some notes on slicing:

- Python, like c (which it was developed in) and some other languages, identifies the first element of an array as element ZERO, not ONE!

- As an additional complication, when slicing, Python does NOT include the last element specified, but stops at the one before it.

# For loops, and avoiding them

We'd like to test Python's plotting capabilities in the notebook. 

Let's plot $y=\sin{x}$ as a function of ("versus") x, for an x array with 500 elements from 0 to $2\pi$, i.e.

    x=np.linspace(0,2*np.pi,5000)

We'll try four different ways to do this, with widely varying speeds.  __First, we need to set up an x array:__

````python
 x=np.linspace(0,2*np.pi,5000)
````

### 1) the least efficient way: 

A) create a list

B) append new elements to the list using its 'append' method (i.e., a function that is part of/attached to the list object)

C) then convert to a 'numpy' array.

We will use the %%timeit ipython magic to see how long each method takes. 

**Use the below box to get help on %%timeit (with '?').  Then, execute the next code box containing the for loop.**

````python
%%timeit?
````

````python
%%timeit
global y  # normally %%timeit uses an internal variable name space so the variable names don't work outside of it.
          # global y makes y still work as a variable outside the %%timeit cell
y_tmp = [] # create an empty list.  Text after a # is treated as a comment in Python and not executed.
for x1 in x: y_tmp.append(np.sin(x1))
y=np.array(y_tmp)
````

__Note:__ For multi-line for loops, you just indent everything that gets executed at each iteration by four spaces.  All code that is indented is then treated as 'inside' the loop till you get to some unindented code.

I.e., the following methods are _all_ equivalent to what we just did.  

Note also that, __other than indents__, extra white space or making code span multiple lines generally makes no difference in python.  __Execute the code in the below code box to check.__

````python
# Let's look at the range of the array we created, to make sure all methods yield identical results.

# We need to do this outside of a %%timeit cell or else it'll print the result many times!

# Check min/max values of y from the previous result.
print(f'Min: {y.min():.4f}, max: {y.max():.4f}')


# Either this:
y_tmp = [] # create an empty list
for x1 in x:
    y_tmp.append(np.sin(x1))
y=np.array(y_tmp)
print(f'Min: {y.min():.4f}, max: {y.max():.4f}')


# or this:
y_tmp = [] # create an empty list
for x1 in x:
    
    y_tmp.append(np.sin(x1))
    +     0

y=np.array(y_tmp)
print(f'Min: {y.min():.4f}, max: {y.max():.4f}')

# BOTH of these yield results identical to that from the code box where we did the timing, 
# above, since the extra space doesn't matter.
````

### 2) Another way to do the same thing, with similar speed: 

A) create an array of the right size

B) use 'enumerate()' to loop over its elements.  

__Use the box above where you got help on '%%timeit' to look at the documentation on 'enumerate''.  Then execute the code box below.__


````python
%%timeit
y = np.zeros(size(x))   # y = x*0. would also work
for i,x1 in enumerate(x): 
    y[i] = np.sin(x1)
````

### 3) Another method of comparable  efficiency, but with more compact code: 

A) Use a "list comprehension" to calculate all the elements of y

B) Then convert to an array 

__Note:__ The `map()` function in Python is an older way to do all this (i.e., to act on all elements of a list at once).

__Execute the below code box__.

````python
%%timeit
y = [np.sin(x1) for x1 in x] 

y = np.array(y)
````

### 4) The best way: use array operations

Numpy functions can generally operate on numpy arrays all at once, and do so in a well-optimized way.  The more adept you get at redesigning your code so that it is done as a set of array operations, the faster your code will be!

# After we've discussed statistics: Generating Random Numbers 

In Python, we can use the `numpy.random.rand()` function to generate random numbers, evenly distributed between 0 & 1.

The syntax is:

    randnum=random.rand(ndim1 , ndim2 , ndim3...)

where ndim1,ndim2, etc. are the size of the 1st, 2nd, etc. dimensions of the output array `randnum` (unlike with `np.zeros`, this is not provided as a tuple; `random.random_sample()` can do the same thing but use tuples with the `size` keyword).  


````python
# Start using the numpy `random` module
# if we did our setup right this is superfluous as %pylab imported all of numpy with no prefix
import numpy.random as random
````

Since `random.rand()` generates random #s evenly distributed between 0 & 1, we can call the result tails if the result is 0-0.5, and heads if it is 0.5-1.

__Repeatedly execute the below code box with shift-enter__.

````python
if random.rand(1) < 0.5:
    print("tails") #note: indented by 4 spaces, no less! See https://www.python.org/dev/peps/pep-0008/
# Indents are used to indicate multi-line 'then' statements that follow an if or else, just as they are used
# for multiline code executed in a for loop
else:
    print("heads")
````

# Creating a function

The function we described in class is in the code cell below:

````python
def sim2coins(ntests):
# simulate ntests tosses of 2 coins 
    coin1=random.rand(ntests) > 0.5
    coin2=random.rand(ntests) > 0.5
    return np.sum( coin1 == coin2 )/ntests
````

__In the box below, test out the function.  E.g.: `print( sim2coins(1000) )` or  `a=sim2coins(1000)`.  Try different numbers of trials, large and small, and see how the values compare to your expectation.__

````python
print(sim2coins(100))
print(sim2coins(1000))
print(sim2coins(10000))
print(sim2coins(100000))
````
